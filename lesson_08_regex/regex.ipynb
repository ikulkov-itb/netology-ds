## Задание 1

Напишите функцию, которая принимает на вход строку и проверяет является ли она валидным транспортным номером (1 буква, 3 цифры, 2 буквы, 2-3 цифры). Обратите внимание, что не все буквы кириллического алфавита используются в транспортных номерах.

Если номер валиден, то функция должна возвращать отдельно номер и регион.

Примеры работы программы:  

1. `car_id = 'A222BC96'`

Результат: 
`Номер A222BС валиден. Регион: 96`

2. `car_id = 'АБ22ВВ193'`

Результат: 
`Номер не валиден`

````python
import re
from re import RegexFlag as fl
````

````python
def regno_check(regno):
    pat = r'([АВЕКМНОРСТУХ](\d{3})[АВЕКМНОРСТУХ]{2})(\d{2,3})'
    # По-хорошему, ещё должен быть словарь регионов, но там слишком много придётся перечислять
    matches = re.findall(pat, regno, flags=fl.I)
    if len(matches) > 0 and int(matches[0][1]) > 0 and int(matches[0][2]) > 0:
        return 'Номер %s валиден. Регион: %d' % (matches[0][0], int(matches[0][2]))
    else:
        return 'Номер не валиден'

print(regno_check('А222ВС96'))
print(regno_check('Б222ВС96'))
print(regno_check('Б000ВС96'))
````

## Задание 3

Напишите функцию, которая будет возвращать акроним по переданной в нее строке со словами.

Примеры работы программы:  

1. `some_words = 'Информационные технологии'`

Результат: 
`ИТ`

2. `some_words = 'Near Field Communication'`

Результат: 
`NFC`

````python
def acronym(text):
    pattern = r'\b([^\s]+?)'
    letters = re.findall(pattern, text)
    return ''.join(letters).upper()

print(acronym('Информационные технологии'))
print(acronym('Near Field Communication'))
````

## Задание 4

Напишите функцию, которая будет принимать на вход список email-адресов и выводить их распределение по доменным зонам.

Пример работы программы:  

`emails = ['test@gmail.com', 'xyz@test.in', 'test@ya.ru', 'xyz@mail.ru', 'xyz@ya.ru', 'xyz@gmail.com']`

Результат: 
```
gmail.com: 2
test.in: 1 
ya.ru: 2 
mail.ru: 1
```

````python
emails = ['test@gmail.com', 'xyz@test.in', 'test@ya.ru', 'xyz@mail.ru', 'xyz@ya.ru', 'xyz@gmail.com']

domain_cnt = {}

for addr in emails:
    pattern = r'([\w\.]+)@([\w+\.]+)'
    parts = re.findall(pattern, addr)
    if parts:
        domain = parts[0][1]
        if domain_cnt.get(domain, ''):
            domain_cnt[domain] += 1
        else:
            domain_cnt[domain] = 1

for k, v in domain_cnt.items():
    print('%s: %d' % (k, v))
````

## Задание 5 (необязательное)

Напишите функцию, которая будет подсчитывать сколько слов начинается на гласные, а сколько на согласные буквы в тексте (текст может быть написан как с использованием букв кириллицы, так и латиницы).

Пример работы программы:  

`some_text = 'Эталонной реализацией Python является интерпретатор CPython, поддерживающий большинство активно используемых платформ. Он распространяется под свободной лицензией Python Software Foundation License, позволяющей использовать его без ограничений в любых приложениях, включая проприетарные.'`

Результат: 
```
Слов на гласные буквы: 9
Слов на согласные буквы: 21 
```

````python
vowel_pattern = r'\b([euioaяиюэоаыуеё]+?)'
consonant_pattern = r'\b((?![euioaяиюэоаыуе])[a-zа-яё]+?)'

some_text = 'Эталонной реализацией Python является интерпретатор CPython, поддерживающий большинство активно используемых платформ. Он распространяется под свободной лицензией Python Software Foundation License, позволяющей использовать его без ограничений в любых приложениях, включая проприетарные.'

vowels = re.findall(vowel_pattern, some_text, flags=fl.I)
consonants = re.findall(consonant_pattern, some_text, flags=fl.I)

print('Слов на гласные буквы: %d' % len(vowels))
print('Слов на согласные буквы: %d' % len(consonants))
````

## Задание 6 (необязательное)
Напишите функцию, которая будет проверять номер сотового телефона на валидность (только для российского формата), если он валиден, то переводить его в формат:   
+7-xxx-xxx-xx-xx  
Постарайтесь предусмотреть как можно больше адекватных форматов изначального ввода номера.
Примеры работы программы:  

1. `phone = '+7 955 555-55-55'`

Результат: 
`+7-950-555-55-55`

2. `phone = '8(955)555-55-55'`

Результат: 
`+7-950-555-55-55`

3. `phone = '+7 955 555 55 55'`

Результат: 
`+7-950-555-55-55`

4. `phone = '7(955) 555-55-55'`

Результат: 
`+7-950-555-55-55`

5. `phone = '423-555-55-5555'`

Результат: 
`Номер не валиден`

6. `phone = '123-456-789'`

Результат: 
`Номер не валиден`

````python
def format_phone(phone):
    check_pattern = r'^\+?(?:7|8)?[\(\s-]*(?:\d{3})[\)\s-]*(?:\d(?:[\s-]*)){7}$'
    if re.findall(check_pattern, phone):
        clean_phone = re.sub(r'(\D)*', '', phone)
        return re.sub(r'(?:7|8)(\d{3})(\d{3})(\d{2})(\d{2})', r'+7-\1-\2-\3-\4', clean_phone)
    else:
        return 'Номер не валиден'

print(format_phone('+7 955 555-55-55'))
print(format_phone('8(955)555-55-55'))
print(format_phone('+7 955 555 55 55'))
print(format_phone('7(955) 555-55-55'))
print(format_phone('423-555-55-5555'))
print(format_phone('123-456-789'))
print(format_phone('8-123-456-789-0'))
````
